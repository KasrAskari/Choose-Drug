## 1- Impoting Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 2- Loading Dataset

```python
df = pd.read_csv("drug200.csv", delimiter=",")
df[0:5]
```

## 3- EDA

```python
# df.describe()
# df.shape
df.info()
```

## 4- Pre-processing

- X datas

```python
X = df[['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K']].values
X[0:5]
```

- y datas

```python
y = df["Drug"]
y[0:5]
```

### - Pre-processing from 'X' Datas

```python
from sklearn import preprocessing

le_sex = preprocessing.LabelEncoder()
le_sex.fit(['F', 'M'])
X[:,1] = le_sex.transform(X[:,1])

le_BP = preprocessing.LabelEncoder()
le_BP.fit(['LOW', 'NORMAL', 'HIGH'])
X[:,2] = le_BP.transform(X[:,2])

le_Chol = preprocessing.LabelEncoder()
le_Chol.fit(['NORMAL', 'HIGH'])
X[:,3] = le_Chol.transform(X[:,3])

X[0:5]
```

## 5- Setting up "DecisionTree"

```python
from sklearn.model_selection import train_test_split

X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.3, random_state=3)
```

- shaping of X_trainset & y_trainset

```python
print('Shape of X_training set {}'.format(X_trainset.shape),'&'
      ,'Size of y_training set {}'.format(y_trainset.shape))
```

- shaping of X_testset & y_testset

```python
print('Shape of X training set {}'.format(X_testset.shape),'&'
      ,'Size of Y training set {}'.format(y_testset.shape))
```

## 6- Modeling DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier

drugTree = DecisionTreeClassifier(criterion='entropy', max_depth=4)
drugTree
```

### - Fitting the Model

```python
drugTree.fit(X_trainset, y_trainset)
```

## 7- Prediction

```python
predTree = drugTree.predict(X_testset)

print(predTree [0:5])
print(y_testset [0:5])
```

## 8- Evaluation

```python
from sklearn import metrics

print("Accuracy:", metrics.accuracy_score(y_testset, predTree))
```
